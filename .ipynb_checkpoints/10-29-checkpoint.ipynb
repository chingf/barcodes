# Helper functions

```python
def relu(X):
    return np.maximum(0, X)
```

```python
def make_bump(size, center):
    """
    Generates an exponential falloff.
    """
    
    assert(center < size)
    x = np.linspace(0, 100)
    halfbump = np.exp(-0.04*x)
    halfbump = halfbump/halfbump.max()
    bump = np.zeros(size)
    bump[:halfbump.size] = halfbump
    bump[-halfbump.size+1:] = halfbump[1:][::-1]
    offset = 0.
    bump = bump - offset
    bump[bump < 0] = 0
    bump /= bump.max()
    bump = np.roll(bump, center)
    return bump

plt.figure(figsize=(4,2))
plt.scatter(np.arange(201), make_bump(201, 28))
plt.title('Example Spatial Input')
plt.show()
```

# Dynamics Function

```python
def get_activity(T, i, softmax_temp=1):
    """
    Runs T timesteps of dynamics given
    constant input i
    """
    
    x = x_prev = np.zeros(N)
    x_prev = i
    activities = []
    for t in range(T):
        x = J@(x_prev) + i
        x[b_start:b_end] = softmax(x[b_start:b_end]*softmax_temp)
        x[b_start:b_end] -= softmax_offset
        x = relu(x)
        x_prev = x
        activities.append(x)
    return activities
```

# Modifiable Parameters

```python
n_spatial = 201
n_barcode = 1001
n_cache = 1
T = 10 # Number of timesteps to simulate
softmax_offset = 1/n_barcode
softmax_action_gain = int(n_barcode/50)
```

```python
seed_locs = [int(n_spatial*0.3), int(n_spatial*0.5)]
print(f'Caches at locations {seed_locs}')
```

# Network initializations

```python
N = n_spatial + n_barcode + n_cache

# Defining useful indices
s_start = 0
s_end = n_spatial
b_start = s_end
b_end = b_start + n_barcode
c_start = b_end
c_end = -1
SEED = c_start
```

```python
def make_J(seed_locs, verbose=True):
    J = np.zeros((N, N)) # (s,b,c) ordering

    # Set sparse spatial-barcode connections
#     for i in range(n_spatial):
#         # One-to-one
#         chosen_bs = np.array([i])
#         J[b_start + i, s_start_i] = 1
        
#         # Randomly select K
#         chosen_bs = np.random.choice(n_barcode, size=2, replace=False)
#         J[b_start + chosen_bs, s_start+i] = 1/chosen_bs.size

#         if verbose and (i in seed_locs):
#             plt.figure(figsize=(1,3))
#             plt.plot()
#             plt.title(f'Spatial cell {i}')
#             plt.xlabel('Connected Barcode Cells')
#             plt.show()
            
    # Random gaussian + decorrelation
    initial_J = np.random.randn(n_spatial, n_barcode)
    mask_J = (np.random.rand(n_spatial, n_barcode) < 0.005).astype(float) # % chance of on
    initial_J = np.multiply(initial_J, mask_J)
    final_J = initial_J.copy()
    for i in range(n_spatial):
        final_J[i] -= 0.5*initial_J[(i-1)%n_spatial]
        final_J[i] -= 0.5*initial_J[(i+1)%n_spatial]
        J[b_start:b_end, s_start + i] = final_J[i]
        if verbose and (i in seed_locs):
            plt.figure(figsize=(3, 1))
            plt.plot(final_J[i])
            plt.title(f'Spatial cell {i}')
            plt.xlabel('Connected Barcode Cells')
            plt.show()
            
    J[s_start:s_end, b_start:b_end] = J[b_start:b_end, s_start:s_end].T
    
    # Set barcode-cache connections
    for l in seed_locs:
        bump = make_bump(n_spatial, l)
        b_activations = J[b_start:b_end, s_start:s_end]@bump
        b_activations = softmax(b_activations*softmax_action_gain) - softmax_offset

        chosen_bs = np.argwhere(J[l, b_start:b_end]>0).flatten()
        #chosen_bs = np.argwhere(b_activations>0).flatten()

        J[b_start + chosen_bs, SEED] = 1
        J[SEED, b_start + chosen_bs] = 1
        if verbose:
            plt.figure(figsize=(3, 1))
            plt.plot(b_activations)
            plt.title(f'Barcode activations for spatial cell {l}')
            plt.show()
            
    return J
```

# Testing decorrelation

```python
temp = softmax_action_gain
```

```python
J = make_J(np.arange(n_spatial), verbose=False)
```

```python
place_input_activity = []
barcode_activity_T1 = []
barcode_activity_T5 = []
barcode_activity_T20 = []
for loc in np.arange(n_spatial):
    _input = np.zeros(N)
    _input[s_start:s_end] = make_bump(n_spatial, loc)
    place_input_activity.append(_input[s_start:s_end])
    activity = get_activity(1, _input, softmax_temp=temp)[-1]
    barcode_activity_T1.append(activity[b_start:b_end])
    activity = get_activity(5, _input, softmax_temp=temp)[-1]
    barcode_activity_T5.append(activity[b_start:b_end])
    activity = get_activity(20, _input, softmax_temp=temp)[-1]
    barcode_activity_T20.append(activity[b_start:b_end])

fig, axs = plt.subplots(1, 4, figsize=(3*4,3))
axs[0].imshow(np.corrcoef(place_input_activity))
axs[0].set_title('Input Correlation')
axs[1].imshow(np.corrcoef(barcode_activity_T1))
axs[1].set_title('Barcode Correlation, t=1')
axs[2].imshow(np.corrcoef(barcode_activity_T5))
axs[2].set_title('Barcode Correlation, t=5')
axs[3].imshow(np.corrcoef(barcode_activity_T20))
axs[3].set_title('Barcode Correlation, t=20')
plt.show()
```
