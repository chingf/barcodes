```python
import matplotlib as mpl
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
num_pts = 16

# Visit points
theta = np.linspace(0, 2 * np.pi, 201)
x = np.cos(theta)
y = np.sin(theta)
pts = np.linspace(0, x.size, num=num_pts, endpoint=False).astype(int)

# Caching points
cache_x = x + np.random.normal(0, 1.5, x.size)
cache_y = y + np.random.normal(0, 1.5, y.size)
cache_z = np.random.normal(0, 0.1, x.size)*0 + 0.05
cache_x -= cache_x.mean()
cache_y -= cache_y.mean()

colors = plt.cm.viridis(np.linspace(0, 1, num_pts))
ax.scatter(x[pts], y[pts], c=colors)
ax.scatter(cache_x[pts], cache_y[pts], cache_z[pts], c=colors)
ax.plot(x, y, 0, linewidth=2, color='gray', label='Visits')
ax.plot(cache_x[pts], cache_y[pts], cache_z[pts], color='red', alpha=0.4, label='Caching')
ax.legend()
ax.set_xticklabels([])
ax.set_yticklabels([])
ax.set_zticklabels([])

plt.show()
```
